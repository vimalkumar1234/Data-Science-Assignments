```python
from statsmodels.tsa.stattools import adfuller

def adf_test(dataset):
  dftest = adfuller(dataset, autolag = 'AIC')
  print("1. ADF : ",dftest[0])
  print("2. P-Value : ", dftest[1])
  print("3. Num Of Lags : ", dftest[2])
  print("4. Num Of Observations Used For ADF Regression and Critical Values Calculation :", dftest[3])
  print("5. Critical Values :")
  for key, val in dftest[4].items():
      print("\t",key, ": ", val)
```

```python
adf_test(data['rate'])
```

```python
from pmdarima import auto_arima 
stepwise_fit= auto_arima(data['rate'],suppress_warnings=True)
stepwise_fit.summary()
```

```python
from statsmodels.tsa.arima.model import ARIMA 
model = ARIMA(data['rate'], order=(4,1,5))
model=model.fit()
model.summary()
```

```python
print(data.shape)
train=data.iloc[:-30]
test=data.iloc[-30:]
print(train.shape,test.shape)
#print(test.iloc[0],test.iloc[-1])
```

```python
start=len(train)
print(start)
end=len(train)+len(test)-1
print(end)
pred=model.predict(start=start,end=end,typ='levels')
print(pred)
```

```python
test['rate'].plot(figsize=(12,5),legend=True)
pred.plot(figsize=(12,5),legend='ARIMA Predictions')
```

```python
data['rate'].mean()
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
rmse=sqrt(mean_squared_error(pred,test['rate']))
print(rmse)
```
