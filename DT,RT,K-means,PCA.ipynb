# PCA

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
```

```python
scale_df = scale.fit_transform(df)
```

```python
from sklearn.decomposition import PCA
pca = PCA(0.90)
```

```python
x_pca = pca.fit_transform(scale_df)
x_pca
```

```python
x_pca.shape
```

```python
df.shape
```

```python
y_dep.shape
```

# TrainTestSplit

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_pca,y_dep,random_state=5,test_size=0.2)
```

```python
print(x_train.shape,y_train.shape,x_test.shape,y_test.shape)
```

# Decision Tree 

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
```

```python
model_dt = dt.fit(x_train,y_train)
```

```python
dt_pred = model_dt.predict(x_test)
```

```python
dt_score = model_dt.score(x_test,y_test)*100
dt_score
```

# Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
rt = RandomForestRegressor()
```

```python
model_rt = rt.fit(x_test,y_test)
```

```python
rt_score = model_rt.score(x_test,y_test)*100
rt_score
```

# Accuracy

```python
acc = pd.DataFrame({'Model':['Decision Tree','Random Forest'],'Accuracy':[dt_score,rt_score]})
acc
```

# k-means clustering

```python
from sklearn.cluster import KMeans
kmeans = KMeans()
```

```python
a=[]
for i in range(1,40):
    kmeans = KMeans(i)
    kmeans.fit(x_pca)
    a.append(kmeans.inertia_)
```
